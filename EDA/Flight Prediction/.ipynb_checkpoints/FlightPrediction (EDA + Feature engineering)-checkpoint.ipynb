```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
## Display all the columns of the dataframe
pd.set_option('display.max_columns',None)
```

```python
dataset = pd.read_excel('Data_Train.xlsx')
dataset.head()
```

```python
dataset.info()
```

```python
dataset['Date'] = dataset['Date_of_Journey'].str.split('/').str[0]
dataset['Month'] = dataset['Date_of_Journey'].str.split('/').str[1]
dataset['Year'] = dataset['Date_of_Journey'].str.split('/').str[2]
dataset.head(2)
```

```python
dataset.info() #Here we can see that date,month and year is of type object
```

```python
#Converting date,month and year to type int
dataset['Date'] = dataset['Date'].astype(int)
dataset['Month'] = dataset['Month'].astype(int)
dataset['Year'] = dataset['Year'].astype(int)
```

```python
dataset.info()
```

```python
#Dropping date_of_journey column as it is of no importance now
dataset.drop('Date_of_Journey',axis = 1, inplace = True)
dataset.head(10)
```

```python
dataset['Arrival_Time'] = dataset['Arrival_Time'].str.split(' ').str[0]
dataset['Arrival_Hour'] = dataset['Arrival_Time'].str.split(':').str[0]
dataset['Arrival_Minute'] = dataset['Arrival_Time'].str.split(':').str[1]
dataset.head(3)
```

```python
dataset.info()
```

```python
dataset['Arrival_Hour'] = dataset['Arrival_Hour'].astype(int)
dataset['Arrival_Minute'] = dataset['Arrival_Minute'].astype(int)
dataset.drop('Arrival_Time',axis = 1, inplace = True)
dataset.head()
```

```python
dataset['Departure_Hour'] = dataset['Dep_Time'].str.split(':').str[0]
dataset['Departure_Minute'] = dataset['Dep_Time'].str.split(':').str[1]
dataset['Departure_Hour'] = dataset['Departure_Hour'].astype(int)
dataset['Departure_Minute'] = dataset['Departure_Minute'].astype(int)
dataset.drop('Dep_Time',axis = 1, inplace = True)
dataset.head()
```

```python
dataset.info()
```

```python
dataset['Total_Stops'].unique()
```

```python
dataset['Total_Stops'] = dataset['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
# we put nan as 1 because the mode of the Total_Stops column is 1 stop. We can see it by,dataset.mode()['Total_Stops'][0]
dataset.head()
```

```python
#since, the route info is given by total_stops, we drop route column
dataset.drop('Route',axis = 1,inplace = True)
dataset.info()
```

```python
dataset = pd.get_dummies(dataset,columns=['Airline','Source','Destination','Additional_Info'],drop_first = True)
```
