```python
#3. write a function which will take input as a dict and find me output as a list of the values even
#in case of two level of nesting
values_list = []
def value(d):
    '''this function will take dict as input and return its values in a single list'''
    for i in d:
        if type(d[i])==dict:
            return value(d[i])
        else:
            values_list.append(d[i])
    return values_list
d = {'k':1,'k1':4,'k2':{'k2a':(1,2,4),'k2b':(4,5,6)}}
value(d)    
```

```python
# 16.write a function to read a complete pdf file
def function(filename):
    import PyPDF2 as pdf
    f = open(filename,'rb')
    f_read=pdf.PdfFileReader(f)
    f_page=f_read.getPage(0)
    print(f_page.extractText())
    f.close()
filename = input("enter the pdf file name")
function(filename)
```
