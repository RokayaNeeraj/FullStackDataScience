```python
import mysql.connector as conn
mydb = conn.connect(host = 'localhost', user = 'root', passwd = 'password')
mydb
```

```python
cursor = mydb.cursor()
```

```python
#cursor.execute("drop database cardataset") # dropping the table because of running it multiple times
cursor.execute("create database cardataset")
cursor.execute("show databases")
cursor.fetchall()
```

```python
#creating table inside the database
cursor.execute("create table cardataset.car(col1 varchar(20),col2 varchar(20), col3 varchar(20),col4 varchar(20),col5 varchar(20),col6 varchar(20),col7 varchar(20))")
mydb.commit()
```

```python
cursor.execute("use cardataset")
cursor.execute("show tables")
cursor.fetchall()
```

```python
#Dumping car.data into the database table
import csv
with open('car.data','r') as f:
    car_data = csv.reader(f,delimiter = '\n')
    for i in car_data:
        row_data = tuple(i[0].split(','))
        query = "insert into cardataset.car values {}".format(row_data)
        cursor.execute(query)
#         cursor.execute("Insert into cardataset.car values ()".format(row_data))
mydb.commit()
```

```python
query = "select * from cardataset.car"
cursor.execute(query)
cursor.fetchall() #displaying the data inside the car table
```

```python
#try to group all the data with col1 and count occurances of each and every record based on col1 value
cursor.execute("select count(col1), col1 from cardataset.car group by col1 ")
cursor.fetchall()
```

```python
#try to filter a record where col3 value will be 4
cursor.execute("select * from cardataset.car where col3 = 4")
cursor.fetchall()
```

```python
#try to update a col3 value with 8 wherever you have value equal to 2
cursor.execute("update cardataset.car set col3='8' where col3= '2'")
cursor.execute("Select * from cardataset.car where col3 = '2'")
cursor.fetchall() # it indicates that we dont have col3 value as '2' as all was changed to 8
```

```python
#try to delete table
cursor.execute("drop table cardataset.car") #dropping table car
cursor.execute("use cardataset")
cursor.execute("show tables")
cursor.fetchall()  #No tables are shown as the table is dropped
```

```python
#try to delete database
cursor.execute("drop database cardataset") #database deleted
cursor.execute("show databases")
cursor.fetchall()
```
