### BOX PLOT

```python
df.describe()
```

```python
df.plot(kind = 'box', figsize = (20,10),color = {'boxes' : 'g','whiskers' : 'r'})
#Q2 or Median(50%) is the middle horizontal line,below that is Q1(25%) and above that is Q3(75%).The range between Q3 and
#Q1 is the interquartile range.
#Median is the middle most value after we sort the numbers. 
#The topmost line(max no) is calculated by Q3 + 1.5(IQR), the bottom most line(min) is calculated by Q1 -1.5(IQR).IQR = Q3-Q1
# The data above Q3 and below Q1 is treated as outlier in boxplot.
```

```python
df.plot(kind = 'box', figsize = (20,10),color = {'boxes' : 'g','whiskers' : 'r'},vert = False)
```

```python
import folium as f
import pandas as pd
```

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/jtemporal/folium-101/main/cities.csv')
df2.head()
```

```python
my_map = f.Map(location = [13.133,16.10],
               zoom_start = 2)
my_map
```

```python
city = df2.loc[0]
city
```

```python
# adding colors to the markers in map
def select_marker_color(row):
    if row['lived'] == 'yes':
        return 'yello'
    elif row['lived'] == 'no' and row['visited'] == 'yes':
        return 'purple'
    return 'blue'
```

```python
df2['color'] = df2.apply(select_marker_color,axis = 1)
df2.head()
```

```python
for _,city in df2.iterrows(): #_ basically means the values are not important and we arent using it further
    
    f.Marker(location = [city['latitude'],city['longitude']],
        popup = city['name'],
        tooltip = city['name'],
        icon = f.Icon(color = city['color'],prefix = 'fa',icon = 'circle')
        ).add_to(my_map)
```

```python
import seaborn as sns
df1 = sns.load_dataset('tips')
df1
```

### plotting smokers vs non-smoker

```python
#use this approach for seaborn as it will be easier
#relplot means relation plot
sns.relplot(x = 'total_bill', y = 'tip',data = df1, hue = 'smoker') #hue helps in colors
```

```python
sns.relplot(x = 'total_bill', y = 'tip',data = df1, style = 'smoker')
```

```python
sns.relplot(x = 'total_bill', y = 'tip',data = df1, style = 'size',hue = 'smoker') #change column name in style and compare 
```

```python
sns.relplot(x = 'total_bill', y = 'tip', kind = 'line', data = df1,hue = 'smoker')
```
