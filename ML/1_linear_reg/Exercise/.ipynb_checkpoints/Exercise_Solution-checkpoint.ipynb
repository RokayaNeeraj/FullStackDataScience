```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

```python
percapita = pd.read_csv("canada_per_capita_income.csv")
percapita
```

```python
plt.xlabel("Year")
plt.ylabel("Prices")
plt.scatter(percapita.year,percapita['per capita income (US$)'],color = 'red')
```

```python
new_df = percapita.drop("per capita income (US$)",axis = 1)
new_df
```

```python
percapita['per_capita_income'] = percapita['per capita income (US$)']
percapita = percapita.drop("per capita income (US$)",axis = 1)
percapita
```

```python
model = linear_model.LinearRegression()
p = model.fit(new_df,percapita.per_capita_income)
```

```python
model.predict([[2020]])
```
